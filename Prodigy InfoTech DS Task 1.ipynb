**Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, such as the distribution of ages or genders in a population.**



**Step 1:** Import the Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

**Step 2:** Load the dataset

```python
data=pd.read_csv('API_SP.POP.TOTL_DS2_en_csv_v2_92.csv',skiprows=4)
# to display entire column
pd.set_option('display.max_columns',None)
# to display entire rows
pd.set_option('display.max_rows',None)
```

**Step 3:** Drop irrelevant columns

```python
data = data.drop(columns=['Unnamed: 68'])
data.shape
```

**Step 4:** Select and clean data for 2020

```python
data_2020 = data[['Country Name', '2020']].dropna()
```

```python
data['2020'] = data['2020'].fillna(0)
```

**Step 5:** Sort data by population

```python
top_countries = data_2020.sort_values(by='2020', ascending=False).head(10)
```

**Step 6:** Plot a Bar Graph

```python
plt.figure(figsize=(12, 6))
plt.bar(top_countries['Country Name'], top_countries['2020'], color='cyan')
plt.title('Top 10 Countries by Population in 2020',fontweight='bold')
plt.xlabel('Country', color='brown',fontsize='14', fontweight='bold')
plt.ylabel('Population in 2020', color='brown', fontsize='14',fontweight='bold')
plt.xticks(rotation=25)
plt.show()
```

## Code for Histogram for the above cleaned data

```python
population_2020 = data_2020['2020']

# Plot the histogram
plt.figure(figsize=(10, 6))
plt.hist(population_2020, bins=10, color='skyblue', edgecolor='black')
plt.title('Population Distribution in 2020', fontweight='bold')
plt.xlabel('Population', color='brown',fontsize='14', fontweight='bold')
plt.ylabel('Frequency', color='brown',fontsize='14', fontweight='bold')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
